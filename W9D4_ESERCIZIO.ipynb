```python
#ESERCIZIO1: stampare ogni carattere della stringa, uno su ogni riga, usando il ciclo WHILE
nome_scuola = "Epicode"
i = 0 #indexing per la stringa e inizializzazione 
lunghezza_parola = 7 #altra inizializzazione

while i < lunghezza_parola:
    lettera =nome_scuola[i]
    print(lettera)
    i += 1 

#ESERCIZIO2: stampare a video tutti i numeri da 0 a 20 utilizzando il ciclo WHILE 
i = 0
lunghezza_numero = 20

while i <= lunghezza_numero: 
    print(i)
    i += 1
print("Finito.")    
```

```python

#ESERCIZIO3: calcolare e stampare tutte le prime 10 potenze di 2 utilizzando il ciclo WHILE
numero = 2 
i = 0
potenza = 10

while i <= potenza:
    print("2 ^", i, "=", numero**i)
    i += 1
print("Finito.") 

#ESERCIZIO4: calcolare e stampare tutte le prime N potenze di 2 utilizzando il ciclo WHILE, domandando all'utente di inserire N
numero = 2 
i = 0
potenze = int(input("Quali potenze di 2 vuoi calcolare?: "))

while i <= potenze:
    print("2 ^", i, "=", numero**i)
    i += 1

```

```python
#ESERCIZIO7: memorizzare e stampare tutti i fattori di un numero dato in input


#ESERCIZIO8: stampare ogni carattere della stringa, uno su ogni riga, utilizzando il ciclo FOR
nome_scuola = "Epicode"

for lettera in nome_scuola:
    print(lettera)

#ESERCIZIO9: calcolare e stampare tutte le prime 10 potenze di 2 utilizzando un ciclo FOR (con range)
numero = 2
i = 0

for i in range(11):
    print("2 ^", i, "=", numero**i)
print("Finito.")

#ESERCIZIO10: calcolare senza stampare le prime N potenze di K, ognuna di esse andrà memorizzata in coda ad una lista
#alla fine stampare la lista risultante, realizzare sia con WHILE sia con FOR

N = 10
K = 3 
i = 0

for i in range(N):
    potenza = K**i
    print(K, "^", i, "=", potenza)

while i < N:
     potenza = K**i
     print(K, "^", i, "=", potenza) 
     i += 1 
```

```python
#ESERCIZIO11: usare un ciclo FOR per calcolare la media dei guadagni e stamparla
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50]
somma = 0
for guadagno in guadagni:
    somma += guadagno
    print("La somma dei guadagni è:", somma ,"La media dei guadagni è: ", guadagno)

numero_guadagni = len(guadagni)
print(somma//numero_guadagni)

#ESERCIZIO12: stampare per ogni parola quante volte appare la lettera "e", considerando sia maiuscola che minuscola
parole = ["Albergo", "Sedia", "Borgo", "Petalo",  "Eremo", "Belvedere", "Semestre", "Esteta", "Sosta", "Orpello", "Abete", "Orologio", "Cesta", "Ermellino"]

for parola in parole:
    numero_e = parola.count("e")
    numero_E = parola.count("E")
    totale_e= numero_e + numero_E
    print(parola , ":il numero delle 'e' è ", totale_e)
```
