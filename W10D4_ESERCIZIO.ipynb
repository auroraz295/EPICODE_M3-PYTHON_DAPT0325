```python
#ESERCIZIO3: SELEZIONARE IL DATASET diabetes.csv, che raccoglie persone con diabete o meno, 
# e il valore di diverse variabili fisiologiche dei pazienti
''' osservare le dimensioni e un'anteprima di cinque righe
    dare un'occhiata ai metadati delle colonne 
    stampare dei descrittori statistici del dataset 
    selezionare i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50 
    qual è la media della pressione sanguigna diastolica per le diverse fasce di età? '''

import pandas as pd
url = "dataset/diabetes.csv"
df = pd.read_csv(url)

print(df.shape) #dimensioni
print(df.head(5)) #anteprima 5 righe

df.info() #panoramica concisa del dataset
df.describe() #statistiche descrittive del dataset
print(df.columns) #nomi colonne

#PRIMA FASCIA D'ETA'
eta1 = df.loc[:, "Age (years)"] >=20 #filtro1 che seleziona l'età sopra i 20 anni
eta1_1 = df.loc[:, "Age (years)"] <=29 #filtro2 che seleziona l'età sotto i 29 anni
fascia1 = eta1 & eta1_1 #fascia d'età che seleziona tra i 20 e i 29 anni

dati_fascia1 = df.loc[fascia1] #dati relativi alla prima fascia 
media_pressione1 = df["Diastolic blood pressure (mm Hg)"].loc[fascia1] #filtro pressione sanguigna in base all'età

#SECONDA FASCIA D'ETA'
eta2 = df.loc[:, "Age (years)"] >=30 
eta2_1 = df.loc[:, "Age (years)"] <=39 
fascia2 = eta2 & eta2_1 

dati_fascia2 = df.loc[fascia2] 
media_pressione2 = df["Diastolic blood pressure (mm Hg)"].loc[fascia2] 

#TERZA FASCIA D'ETA'
eta3 = df.loc[:, "Age (years)"] >=40 
eta3_1 = df.loc[:, "Age (years)"] <=50
fascia3 = eta3 & eta3_1 

dati_fascia3 = df.loc[fascia3] 
media_pressione3 = df["Diastolic blood pressure (mm Hg)"].loc[fascia3] 

#STAMPE DATI E MEDIE
print(dati_fascia1.sort_values("Age (years)", ascending=True)) #FASCIA 1
print(media_pressione1.describe())

print(dati_fascia2.sort_values("Age (years)", ascending=True)) #FASCIA 2
print(media_pressione2.describe()) 

print(dati_fascia3.sort_values("Age (years)", ascending=True)) #FASCIA 3
print(media_pressione3.describe()) 

```

```python
#ESERCIZIO5: SELEZIONARE IL DATASET pokemon.csv
''' quali sono i Pokémon leggendari? 
    e quali sono i leggendari di tipo 1 Grass? 
    e leggendari di tipo 1 Ice o Fire? 
    trasformare la colonna Name nell'indice 
    quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
'''

filtro_legendary = df.loc[:, "Legendary"] == True
df["Name"].loc[filtro_legendary]

type1_g = df.loc[:, "Type 1"] == "Grass"
filtro_l_g = filtro_legendary & type1_g

df["Name"].loc[filtro_l_g]

type1_i = df.loc[:, "Type 1"] == "Ice"
type1_f = df.loc[:, "Type 1"] == "Fire"
filtro_i_f = type1_i | type1_f
filtro_l_i_f = filtro_legendary & filtro_i_f
df["Name"].loc[filtro_l_i_f]

df.set_index("Name")

filtro_generation = df.loc[:, "Generation"] == 1
filtro_attacco = df.loc[:, "Attack"] >50
filtro_hp = df.loc[:, "HP"] <60
filtro_finale = filtro_generation & filtro_attacco & filtro_hp

df["Name"].loc[filtro_finale]
```
