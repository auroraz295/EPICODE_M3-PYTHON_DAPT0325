```python
#ESERCIZIO4: l'azienda Object SpA ha creato uan lista di quanti oggetti ha venduto ogni mese nell'ultimo anno
'''lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350]'''
#trasformala in un array NumPy (casting):
'''lst =np.array(lst)'''

#Rispondere alle domande:
#Qual è stata la vendita massima mensile? E quella minima? 
''' MAX: 8150 - MIN: 1280 '''

#Quali sono le vendite mensili maggiori di 5000 oggetti? E quante sono?
''' 5500 7200 6410 8150 7100 5350, sono 6 '''

#Quali sono le vendite minori di 3000 oggetti?
''' 2000 1280 1900 2500, sono 4 ''' 

#In media quanti oggetti sono stati venduti al mese?
''' 4634.16 '''


import numpy as np

lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350]
lst =np.array(lst)
max_5000 = lst > 5000
dati_5000 = lst[max_5000]
min_3000 = lst < 3000

print("Vendita massima mensile:", lst.max())
print("Vendita minima mensile:", lst.min())
print("Vendite mensili maggiori di 5000:", lst[max_5000])
print("Numero di vendite maggiori di 5000:", len(dati_5000))
print("Vendite mensil minori di 3000:", lst[min_3000])
print("Media oggetti venduti:", lst.mean())
```

```python
#ESERCIZIO5: considerando il dizionario fatturati_dict e la Series fatturati_series, accendendo alle stesse informazioni nello stesso modo
#Qual è la differenza tra i due tipi di dato? Cosa si potrebbe fare con la Series che non si può fare con il dizionario?
''' il dizionario non supporta l'indexing, è una struttura non ordinata
    la series supporta l'indexing ed è una struttura ordinata '''

import pandas as pd

fatturati_dict = {1997: 12_000, 1998: 15_000, 
                  1999: 20_000, 2000: 23_000, 
                  2001: 25_000, 2002: 17_000, 
                  2003: 14_000, 2004: 21_000}

fatturati_series = pd.Series([12_000, 15_000, 20_000, 23_000, 25_000, 17_000, 14_000, 21_000], 
                             index=range(1997, 2005))

print(fatturati_dict[1997])
print(fatturati_series[1997])
```
